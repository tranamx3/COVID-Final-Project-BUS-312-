## III. Data Merging

We'll merge both datasdets "all-states-history.csv" and "us_state_vaccinations.csv" to make correlation interpretation of covid cases and vaccination records between the states. 

```python
df1=pd.read_csv("us_state_vaccinations.csv")
```

```python
#step 1
df1=df1.rename(columns={"location":"state"})
df1
```

#### ^In order to merge both dataset we need to find the primary key in the dataset "us_state_vaccinations".  Once we've identified the key being "location" and "date", we have to rename the pk "location" to "state" so that it matches with other primary key from the dataset "all-states-history.csv" this will allow us to merge both dataset together. 

```python
#step 2
df1=df1.replace({"California":"CA",
              "Texas":"TX","Arizona":"AZ","Alabama":"AL","Florida":"FL",
                "Colorado":"CO","New York":"NY","Hawaii":"HI"})
df1
```

#### ^We then replaced the names of the entries in the state variable therefore the entries will match with the other dataset; for example "California" is now "CA".

```python
#step 3
df3=pd.merge(df, df1, on=["state", "date"])
df3
df3.drop(columns=['total_boosters', 'total_boosters_per_hundred'])
```

#### ^We then use the command to merge the primary key "state" and "date" together. This gives a new table that has all the variables from both datasets. We also had to remove two columns from the table "total_boosters" and "total_boosters_per_hundred" since those variables didn't hold any entries.

```python
total1=(df3.groupby("state")["total_vaccinations_per_hundred"].count())
total1
```

#### ^We then made another table that includes all the states that were renamed from the table to make a count table.

```python
plt.pie(total1.values,labels=total1.index,autopct='%.0f%%')
plt.title("Total Vaccination Per Hundred")
plt.savefig("total.pdf",dpi=100)
```

#### ^We then went from that table to make a pie chart that shows the total number of vaccinations (by hundred) in percentage between the states. Noticable here California has the highest percentage by 1% higher whereas all the other states have the same.

```python
g=sns.lmplot(x="death",y="total_distributed",hue="state",data=df3)
g.set(xlim=0,ylim=0,title="total distribution and fully vaccinated")
```

#### ^This scatterplot shows the relations between people fully vaccianted against total distribution (as in state vaccine distribution). The graph shows distinctly that there is a positive correlation where more distribution of vaccine are given between states gives higher number of people fully vaccinated.

```python
g=sns.lmplot(x="people_vaccinated_per_hundred",y="people_fully_vaccinated",hue="state",data=df3)
g.set(xlim=0, ylim=0, title="number of death")
```

#### ^This scatterplot shows the relations between people fully vaccianted against people vaccinated (per hundred). This graph is meant to show that people who are vaccinated also tend to want to be fully vaccinated as well.

```python
g=sns.lmplot(x="people_fully_vaccinated",y="positive",hue="state",data=df3)
g.set(xlim=0,ylim=0,title="people vaccinated and positive")
```

#### ^This is another scatterplot that shows people who are vaccinated and positive but are fully vaccinated as well. Although it's hard to interprete we can see that people who are positive from covid-19 are likely as well to get vaccinated fully.

```python
g=sns.lmplot(x="people_fully_vaccinated",y="negative",hue="state",data=df3)
g.set(xlim=0,ylim=0,title="people vaccinated and negative")
```

#### ^ On the other hand this scatterplot shows people who are vaccinated and negative. This indicates that people who are tested negative still go to be fully vaccinated. We can see that Florida surprisingly has the highest count.
